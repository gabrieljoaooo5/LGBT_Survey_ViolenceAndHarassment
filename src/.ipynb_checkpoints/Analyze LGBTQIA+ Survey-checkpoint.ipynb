## What are the most dangerous places for LGBTQIA+ people?

```python
def filterQuestions(keyword):
    questions = df.question_label.unique().tolist()
    keyword_list = []
    
    for question in questions:
        if keyword in question:
            keyword_list.append(question)
            
    return keyword_list

def filterDataByQuestions(df, questions_list):
    df_questions = df[df.question_label.isin(questions_list)]
    df_questions = df_questions[df_questions.percentage != ':']
    
    df_questions.percentage = df_questions.percentage.astype(int)
    
    return df_questions

def popularAnswers(df):
    countries_list = df.CountryCode.unique().tolist()
    subset_list = df.subset.unique().tolist()
    
    for subset in subset_list:
        df_subset = df[df.subset == subset]
        answers_dict = { }
            
        for country in countries_list:
            df_country = df_subset[df_subset.CountryCode == country]
            df_country = df_country.groupby('answer').sum().sort_values(by='percentage', ascending=False)
            df_country = df_country[df_country.percentage != 0]

            answers_list = df_country.index.tolist()

            for answer_1 in answers_list:
                for answer_2 in answers_list:
                    similarity = SequenceMatcher(a=answer_1,b=answer_2).ratio()
                    if(similarity > 0.6 and similarity != 1):
                        percentage = df_country[df_country.index == answer_2]['percentage'].tolist()[0]
                        df_country[df_country.index == answer_1] += percentage
                        answers_list.remove(answer_2)
                        df_country.drop(answer_2, inplace=True)
                        
            popular_answers = df_country.sort_values('percentage', ascending=False).head(3).index.tolist()
            for answer in popular_answers:
                if answer in answers_dict:
                    answers_dict[answer] += 1
                else:
                    answers_dict[answer] = 1
        
        answers_list = list(answers_dict.keys())          
        for answer_1 in answers_list:
            for answer_2 in answers_list:
                similarity = SequenceMatcher(a=answer_1,b=answer_2).ratio()
                if(similarity > 0.6 and similarity != 1 and answer_1 in answers_dict and answer_2 in answers_dict):
                    answers_dict[answer_1] += answers_dict[answer_2]
                    answers_dict.pop(answer_2, None)
        c = Counter(answers_dict)
        print("*"*50)
        print('\nSubset: ' + subset)
        print(c.most_common(3))
        
def checkSimilarity(answers_list):
    answers_list = answers_list.sort()
    similarity_dict = {}
    for answer_1 in answers_list:
        for answer_2 in answers_list:
            similarity = SequenceMatcher(a=answer_1,b=answer_2).ratio()
            if(similarity > 0.6 and similarity != 1): 
                if answer_1 in similarity_dict:
                    similarity_dict[answer_1] += [answer_2]
                else:
                    similarity_dict[answer_1] = [answer_2]
                answers_list.remove(answer_2)
    return similarity_dict

def mostPopularAnswer(df):
    subset_list = df.subset.unique().tolist()
    result_dict = { }
    for subset in subset_list:
        df_subset = df[df.subset == subset]
        countries_list = df_subset.CountryCode.unique().tolist()
        answers_dict = { }      
        for country in countries_list:
            df_country = df_subset[df_subset.CountryCode == country]
            df_country = df_country.groupby('answer').sum().sort_values(by='percentage', ascending=False)
            df_country = df_country[df_country.percentage != 0]
            
            similar_answers = checkSimilarity(df_country.index.unique().tolist())
            if len(similar_answers) != 0:
                for key, value_list in similar_answers.items():
                    for value in value_list:
                        percentage = df_country[df_country.index == value]['percentage'].tolist()[0]
                        df_country[df_country.index == key] += percentage
                        df_country.drop(value, inplace=True)
                        
            
            popular_answer = df_country.sort_values('percentage', ascending=False).index.tolist()[0]
            if popular_answer in answers_dict:
                answers_dict[popular_answer] += 1
            else:
                answers_dict[popular_answer] = 1
        
        result_dict[subset] = answers_dict
   
    return result_dict

def prepareVisualization(answers_dict):
    answers_list = []
    percentage_list = []
    # convert to percentage
    for answer in answers_dict.values():
        total_countries = sum(answer.values())
        for key, value in answer.items():
            answer[key] = round(value/total_countries*100, 2)
            if (key not in answers_list):
                answers_list.append(key)
            
    subset_list = list(answers_dict.keys()) * len(answers_list)        
    answers_list *= len(answers_dict.keys())
    
    count = 0
    while count < len(answers_list):
        subset = subset_list[count]
        answer = answers_list[count]
        if answer in answers_dict[subset]:
            percentage_list.append(answers_dict[subset][answer])
        else:
            percentage_list.append(0)
        count+=1
        
    return pd.DataFrame.from_dict({'subset': subset_list, 'answers': answers_list, 'percentage': percentage_list})
    
```

```python
def checkSimilarity(answers_list):
    answers_list.sort()
    similarity_dict = {}
    for answer_1 in answers_list:
        for answer_2 in answers_list:
            similarity = SequenceMatcher(a=answer_1,b=answer_2).ratio()
            if(similarity > 0.6 and similarity != 1): 
                if answer_1 in similarity_dict:
                    similarity_dict[answer_1] += [answer_2]
                else:
                    similarity_dict[answer_1] = [answer_2]
                answers_list.remove(answer_2)
    return similarity_dict

def mostPopularAnswer(df):
    subset_list = df.subset.unique().tolist()
    result_dict = { }
    for subset in subset_list:
        df_subset = df[df.subset == subset]
        countries_list = df_subset.CountryCode.unique().tolist()
        answers_dict = { }      
        for country in countries_list:
            df_country = df_subset[df_subset.CountryCode == country]
            df_country = df_country.groupby('answer').sum().sort_values(by='percentage', ascending=False)
            df_country = df_country[df_country.percentage != 0]
            
            similar_answers = checkSimilarity(df_country.index.unique().tolist())
            if len(similar_answers) != 0:
                for key, value_list in similar_answers.items():
                    for value in value_list:
                        percentage = df_country[df_country.index == value]['percentage'].tolist()[0]
                        df_country[df_country.index == key] += percentage
                        df_country.drop(value, inplace=True)
                        
            
            popular_answer = df_country.sort_values('percentage', ascending=False).index.tolist()[0]
            if popular_answer in answers_dict:
                answers_dict[popular_answer] += 1
            else:
                answers_dict[popular_answer] = 1
        
        result_dict[subset] = answers_dict
   
    return result_dict

def prepareVisualization(answers_dict):
    answers_list = []
    percentage_list = []
    # convert to percentage
    for answer in answers_dict.values():
        total_countries = sum(answer.values())
        for key, value in answer.items():
            answer[key] = round(value/total_countries*100, 2)
            if (key not in answers_list):
                answers_list.append(key)
            
    subset_list = list(answers_dict.keys()) * len(answers_list)        
    answers_list *= len(answers_dict.keys())
    
    count = 0
    while count < len(answers_list):
        subset = subset_list[count]
        answer = answers_list[count]
        if answer in answers_dict[subset]:
            percentage_list.append(answers_dict[subset][answer])
        else:
            percentage_list.append(0)
        count+=1
        
    return pd.DataFrame.from_dict({'subset': subset_list, 'answers': answers_list, 'percentage of countries': percentage_list})
    
```

```python
place_questions = filterQuestions('Where')

df_place = filterDataByQuestions(df, place_questions)
answers_dict = mostPopularAnswer(df_place)
df_answers = prepareVisualization(answers_dict)
df_answers.loc[df_answers.answers == 'A street, square, car parking lot or other public place', 'answers'] = 'Public place'

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x="subset", y="percentage of countries", hue="answers", data=df_answers).set_title('Where did it happen?', fontdict = {'fontsize': 20})
plt.savefig('place_plot.png')
```
